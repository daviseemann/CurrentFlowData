```python
import pandas as pd
import requests
import os
```

```python
url_base = 'https://ons-aws-prod-opendata.s3.amazonaws.com/dataset/carga_energia_di/'
years = range(2000, 2001)
urls = [f'CARGA_ENERGIA_{year}.csv' for year in years]
```

```python
def download_csv(url, url_base = 'https://ons-aws-prod-opendata.s3.amazonaws.com/dataset/carga_energia_di/'):

    response = requests.get(f'{url_base}{url}')
    response.raise_for_status()  
    if response.status_code == 200:
        # Save the content to the file
        with open(f'./csv_files/{url}', 'wb') as file:
            file.write(response.content)
        print(f"File saved successfully at {url}")
    
download_csv(urls[0]) # Download the file
```

```python
pd.read_csv('csv_files.csv', sep=';', decimal=';')
```

```python

data_frames = []
for url in urls:
    filename = f'./csv_files/{url}'
    if not os.path.exists(filename):
        try:
            response = requests.get(url)
            response.raise_for_status()
            df = pd.read_csv(response.text)
            df.to_csv(filename, index=False)
            data_frames.append(df)
            print(f"Successfully downloaded data for {url}")
        except Exception as e:
            print(f"Failed to download data for {url}: {e}")
    else:
        print(f"File {filename} already exists. Skipping download.")
```

```python
if data_frames:
    combined_df = pd.concat(data_frames, ignore_index=True)
    print(combined_df.head())
else:
    print("No data frames to concatenate.")

print(combined_df.head(-1))
```
